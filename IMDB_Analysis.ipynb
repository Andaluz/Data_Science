Let's begin with importing our needed libraries.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

Now, let's load in the IMDB dataset.

```python
movies = pd.read_csv('./data/imdb_1000.csv')
```

Here, we will be conducting exploratory data analysis on IMDB's movie data. Let's check out the data.

```python
movies.head()
```

```python
movies.tail()
```

```python
movies.info()
```

```python
# Let's calculate the average movie duration #
print('Average movie duration:', movies.duration.mean())
```

```python
# Let's sort the data by duration #
movies.sort_values('duration')
```

```python
# We can achieve the results from the previous cell using another method #
disp_cols = ['title', 'duration']
movies_sorted = movies.sort_values('duration', ascending=False)

print('Longest Duration')
print(movies_sorted[disp_cols].iloc[:3])

print('---'*20)

print('Shortest Duration')
print(movies_sorted[disp_cols].iloc[-3:])
```

```python
# Let's create a histogram of durations #
movies.duration.plot(kind='hist', bins=int(round(np.sqrt(len(movies)))));
print(movies.duration.describe())
```

```python
# Let's visualize the same result for the previous cell with a box plot #
movies.duration.plot(kind='box', sym='k.');
```

```python
# Let's check approximately how many movies are outliers in our data #
print('Number of outliers in dataset:', movies[movies.duration > 175].title.value_counts().sum())
```

```python
# Let's find out how many movies are in each content rating #
movies.content_rating.value_counts()
```

```python
# Let's visualize the previous cell #
movies.content_rating.value_counts().plot(kind='barh');
plt.title('Count of Movie Ratings');
plt.xlabel('Count of Movies');
plt.ylabel('Movie Rating');
```

```python
# Let's group GP, PASSED, APPROVED, and NOT RATED into the content rating UNRATED #
movies.content_rating.replace(['GP', 'PASSED', 'APPROVED', 'NOT RATED'], 'UNRATED', inplace=True)
movies.content_rating.value_counts()
```

```python
# Let's also group X and TV-MA into the content rating NC-17 #
movies.content_rating.replace(['X', 'TV-MA'], 'NC-17', inplace=True)
movies.content_rating.value_counts()
```

```python
# Earlier, we saw that there were null values in the content_rating column #
movies.isnull().sum()
```

```python
# Let's check out these null values #
movies[movies.content_rating.isnull()]
```

```python
# Let's fill in these null values with UNRATED #
movies.content_rating.fillna('UNRATED', inplace=True)
movies.content_rating.value_counts(dropna=False)
```

```python
# Let's visualize the movie ratings again, but with the cleaned up data this time #
movies.content_rating.value_counts().plot(kind='barh');
plt.title('Count of Movie Ratings');
plt.xlabel('Count of Movies');
plt.ylabel('Movie Rating');
```

```python
# Let's calculate the average star rating for movies 2 hours or longer, and #
# compare that with the average star rating for movies shorter than 2 hours #
movies_long = movies[movies.duration >= 120].star_rating.mean()
movies_short = movies[movies.duration < 120].star_rating.mean()

print('The average star rating for movies 2 hours or longer:', movies_long)
print('The average star rating for movies shorter than 2 hours:', movies_short)
```

```python
# Let's see if we can visualize whether there is a relationship between duration and star rating #
sns.heatmap(movies.corr(), annot=True, cmap='Reds');
sns.pairplot(movies);
sns.lmplot(data=movies, x='duration', y='star_rating');
```

```python
# Let's calculate the average duration for each genre #
movies.groupby('genre').duration.mean()
```

```python
# Let's see if we can visualize whether there is a relationship between duration and content rating #
movies.boxplot(column='duration', by='content_rating', figsize=(10, 5), sym='k.');
plt.xlabel('Content Rating');
plt.ylabel('Duration');

movies.duration.hist(by=movies.content_rating, sharex=True, sharey=True, figsize=(9, 9));
plt.tight_layout();
```

```python
# Let's determine the top rated movie (by star rating) for each genre #
movies.sort_values('star_rating', ascending=False).groupby('genre').title.first()
```

```python
# We can also obtain the same results as the previous cell by iterating through each genre #
# Note: this method can show us movies in which are 'tied' as top rated for the genre #
top_rated = movies.groupby('genre').star_rating.max()
genre_list = movies.sort_values('genre').genre.unique().tolist()

for g in genre_list:
    print('Top Movie(s) for {} Genre: {}'.format(g, movies[(movies.star_rating == top_rated[g]) & (movies.genre == g)].title.tolist()))
```

```python
# Let's check if there are multiple movies with the same title #
duplicates = movies[movies.title.duplicated(False)].copy()
duplicates
```

```python
# Let's calculate the average star rating for each genre (only if it contains at least 10 movies) #
movies.groupby('genre').star_rating.mean()[movies.genre.value_counts() >= 10]
```

```python
# We can also obtain the same results as the previous cell by filtering the count values for each genre #
genres = movies.groupby('genre').star_rating.agg(['mean', 'count'])
genres[genres['count'] >= 10]
```

```python
# The highest rated movie in the dataset contains the actors: Tim Robbins, Morgan Freeman, Bob Gunton #
actor_ratings = movies.groupby('actors_list')['star_rating'].mean()
actor_ratings.sort_values(ascending=False).head()
```

```python
# Tim Robbins seems to only make R-rated movies #
timR = movies[movies.actors_list.str.contains('Tim Robbins')].star_rating.mean()
print('Average star rating of movies involving Tim Robbins:', timR)
movies[movies.actors_list.str.contains('Tim Robbins')]
```

```python
# Morgan Freeman seems to be well-rounded on content_rating but leans towards R-rated movies #
morganF = movies[movies.actors_list.str.contains('Morgan Freeman')].star_rating.mean()
print('Average star rating of movies involving Morgan Freeman:', morganF)
movies[movies.actors_list.str.contains('Morgan Freeman')]
```

```python
# Bob Gunton only appears in one film within this dataset #
# This could imply he was an upcoming actor at this time #
bobG = movies[movies.actors_list.str.contains('Bob Gunton')].star_rating.mean()
print('Average star rating of movies involving Bob Gunton:', bobG)
movies[movies.actors_list.str.contains('Bob Gunton')]
```
